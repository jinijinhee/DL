### 1.3.1 간단한 구현

#### AND gate

```python
# 2.3.1.1 AND 게이트의 구현
def AND (x1, x2) :
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1 * w1 + x2 * w2
    if tmp <= theta :
        return 0
    elif tmp > theta :
        return 1
```

```python
# 2.3.1.2
AND (0,0)  # 0
```

```python
# 2.3.1.3
AND (0,1)  # 0
```

```python
# 2.3.1.4
AND (0,1)  # 0
```

```python
# 2.3.1.5
AND (1,1)  # 1
```

### 1.3.3 가중치와 편향을 반영한 구현

#### AND gate

```python
# 2.3.3.1 가중치와 편향을 도입한 AND 게이트 구현
def AND (x1, x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b= -0.7
    
    tmp =np.sum(x*w)+b
    if tmp <=0:
      return 0
    else :
      return 1
```

```python
# 2.3.3.2
AND (0,0) , AND (0,1) , AND(1,0), AND(1,1)
```

#### NAND gate

```python
# 2.3.3.3 NAND 게이트의 구현 -가중치 편향 부호 바꾸기
def NAND (x1, x2):
    x = np.array([x1,x2])
    w = np.array([-0.5,-0.5])
    b= 0.7
    
    tmp =np.sum(x*w)+b
    if tmp <=0:
      return 0
    else :
      return 1
```

```python
# 2.3.3.4 
NAND (0,0) , NAND (0,1) , NAND(1,0), NAND(1,1)
```

####OR gate

```python
# 2.3.3.5 OR 게이트의 구현 - 편향만 바꿔주기
def OR (x1, x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b= -0.3
    
    tmp =np.sum(x*w)+b
    if tmp <=0:
      return 0
    else :
      return 1
```

```python
# 2.3.3.6
OR (0,0) , OR (0,1) , OR(1,0), OR(1,1)
```

#### XOR gate

```python
# 1.5.2.1 XOR 게이트의 구현
def XOR (x1,x2) :
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)
    y = AND(s1,s2)
    return y
```

```python
# 1.5.2.2 XOR 게이트 축약형
def XOR (x1,x2) :
    return  AND( NAND(x1,x2), OR(x1,x2))
```

```python
# 1.5.2.3
XOR (0,0) # 0
```

```python
# 1.5.2.4
XOR (1,0) # 1
```

```python
# 1.5.2.5
XOR (0,1) # 0
```
